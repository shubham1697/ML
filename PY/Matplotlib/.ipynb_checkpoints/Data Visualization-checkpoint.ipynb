# Install matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(10)
x = x**2
x2 = x**3
print(x)
print(x2)
```

```python
plt.plot(x)
plt.plot(x2)
plt.show()
```

```python
plt.scatter(x,x*2, marker = '^',color = 'orange', label = 'Stock')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Predictions")
plt.show()
```

# More on Plotting

```python
x = np.array([1,2,3,4,5])
y = np.array([10,20,30,40,50])

plt.plot(x,x**2,label = '2018')
plt.plot(x,x**3,label = '2017')
plt.xlabel("No. of Days")
plt.ylabel("Prices")
plt.title("Stock Price Comparison")
plt.legend()
plt.show()
```

# Bar Chart

```python
numbers = np.random.randint(0,10,5)
indices = np.arange(5)

numbers2 = np.random.randint(0,10,5)
indices2 = np.arange(5)

print(numbers)
print(indices)
```

```python
plt.bar(indices,numbers, 0.25,color = 'orange', label = 'Gold')
plt.bar(indices2+0.25,numbers2,0.25, color = 'Green', label = 'Silver')
plt.legend()
plt.show()
```

# Pie Chart

```python
labels = ['Maths','Science','Eng','Hindi']
marks = [80,85,90,60]

plt.pie(marks, labels = labels, radius = 1,explode=(0,0,0.5,0),shadow = 'True')
plt.show()
```
