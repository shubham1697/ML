# Install matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(10)
x = x**2
x2 = x**3
print(x)
print(x2)
```

```python
plt.plot(x)
plt.plot(x2)
plt.show()
```

```python
plt.scatter(x,x*2, marker = '^',color = 'orange', label = 'Stock')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Predictions")
plt.show()
```
